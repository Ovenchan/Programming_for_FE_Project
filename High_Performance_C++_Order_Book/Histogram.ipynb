# Insert

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

no_tuning = pd.read_csv("./vector_without_tunning/latency_insert_vector_without_tuning.csv", header=None)[0].to_numpy()
vector = pd.read_csv("./vector/latency_insert_vector.csv",  header=None)[0].to_numpy()
heaps= pd.read_csv("./stl_heaps/latency_insert_heaps.csv", header=None)[0].to_numpy()

all_x = np.concatenate([no_tuning, vector, heaps])
bins = np.histogram_bin_edges(all_x, bins=100)

h1, _ = np.histogram(no_tuning,  bins=bins, density=True)
h2, _ = np.histogram(vector,  bins=bins, density=True)
h3, _ = np.histogram(heaps, bins=bins, density=True)
x   = bins[:-1]
w   = np.diff(bins)

plt.bar(x,           h1, width=w/3, align='edge', label='Without Tunning')
plt.bar(x+w/3,       h2, width=w/3, align='edge', label='Vector')
plt.bar(x+2*w/3,     h3, width=w/3, align='edge', label='Heaps')
plt.xlabel('Latency (ns)'); plt.ylabel('Density'); plt.title('Latency Histograms (Insert)')
plt.legend()
plt.tight_layout()
plt.show()
```
