```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv('orders.csv')
# keep the last price of certain Timestamp to form tick price
orders = orders.drop_duplicates(subset=['Timestamp'], keep='last')
orders
```

```python
last_100_orders = orders[-100:]
```

```python
last_100_orders.describe()
```

```python
# Price changes with Timestamp
plt.figure(figsize=(10, 6), dpi=150)
plt.plot(last_100_orders['Timestamp'], last_100_orders['Price'])
plt.xlim(last_100_orders['Timestamp'].min(), last_100_orders['Timestamp'].max())
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.title('Price changes with Timestamp')
plt.show()
```
